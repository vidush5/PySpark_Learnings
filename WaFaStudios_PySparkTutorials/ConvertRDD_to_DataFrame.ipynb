```python
import findspark
findspark.init()

import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.types import StringType, StructField, StructType, IntegerType, ArrayType
from pyspark.sql.functions import col, lit, array, explode, map_keys, map_values
from pyspark.sql import Row

spark = SparkSession.builder.appName("show").getOrCreate()
```

```python
data = [(1, 'maheer'), (2, 'wafa')]
```

```python
rdd = spark.sparkContext.parallelize(data)
```

```python
print(rdd.collect())
```

```python
df = rdd.toDF(['id', 'name'])
df.show()
```

```python
df2 = spark.createDataFrame(rdd, ['id', 'name'])
df2.show()
```
